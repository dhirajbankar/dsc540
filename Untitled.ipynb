```python
import numpy as np
import pandas as pd


df1 = pd.DataFrame({
'A':[1,2,3,4],
'B':[True,False,True,True],
'C':['C1','C2','C3','C4']
})
df2 = pd.DataFrame({
'A':[5,7,8,5],
'B':[False,False,True,False],
'C':['C1','C3','C5','C8'],
},index=[2,3,4,5])
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, on = 'C' , how='inner')
```

```python
df1.merge(df2, on = 'C' , how='outer')
```

```python
df1.merge(df2, on = 'C' , how='left')
```

```python
df1.merge(df2, on = 'C', how='right')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.xlabel("angle")
plt.ylabel("sine & cos")
plt.title('sine wave and cos wave')
plt.grid(True, which='both')

plt.show()
```
