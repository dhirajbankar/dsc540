```python
import requests
from bs4 import BeautifulSoup

URL = 'https://www.worldometers.info/coronavirus/country/us/'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup)
```

```python
python_jobs = soup.find(id="usa_table_countries_today")
```
