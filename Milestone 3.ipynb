## 1) Replace Headers

```python
# Change the column names 
scraped_df.columns= ['A','State','Cases','Deaths','Recovered' , 'Hospitalized' , 'Ref', '7_levele_1','8_levele_1','9_levele_1']
```

```python
# Print newly created datafram with updated headers
scraped_df
```

# 2)  Identify bad data and remove columns which has NaN 

```python
#Drop column A as it has all NaN values
clean_columns_df = scraped_df.dropna(axis=1)
```

```python
# Print dataframe after deleting first column
clean_columns_df
```

# 3) Add a date column to the dataframe

```python
# Add data field in dataframe
clean_columns_df["Updated_date"] = pd.date_range(start='1/13/2021' , end='1/13/2021', periods=len(clean_columns_df))
```

```python
# Print update dataframe 
clean_columns_df
```

# 4) Delete row 56 and 57 as that information is not usefull 

```python
# Clena the footer rows from dataframe 
clean_columns_df = clean_columns_df[0:56 : ]
```

```python
# Print update dataframe 
clean_columns_df
```

```python
# Delete reference column 
del clean_columns_df['Ref']
```
