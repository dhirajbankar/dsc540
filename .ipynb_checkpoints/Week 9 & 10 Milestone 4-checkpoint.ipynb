## 5) Add new date column. 
As we have this data for year 2012 and 2013. So lets convert or add the 8 years in data so we will get 2020 and 2021 data. This data will not be exact but it will be approximatly correct.  

```python
df_copy = df_temperature.copy()
df_copy.head()
```

```python
# As we have this data for year 2012 and 2013. So lets convert or add the 8 years in data so we will get 2020 
# and 2021 data. This data will not be exact but it will be approximatly correct.

from datetime import datetime, date, time, timezone, timedelta
df_copy['Date'] = df_copy['Date'].astype('datetime64[ns]') + timedelta(days= (365*8)) 
df_copy.head()
```
