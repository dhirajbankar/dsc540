```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline
```

```python
im = plt.imread("./DSC_0299.jpg")
im.shape
```

```python
im
```

```python
plt.imshow(im)
plt.show()
```

```python
# Let focus on only face of the image 
im1 = im[300:2000,2000:4000,:]    


plt.imshow(im1)
plt.show()
```

```python
#Let remove alternate columns from Matrix
#Colours
#Each pixel of the image is represented by three integers: 
# the RGB value of its colour. Splitting the image into seperate colour components is just a matter of pulling out the correct slice of the image array:

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

for c, ax in zip(range(3), axs):
    tmp_im = np.zeros(im1.shape, dtype="uint8")
    tmp_im[:,:,c] = im1[:,:,c]
    ax.imshow(tmp_im)
    ax.set_axis_off()
    
```
